```python
#Import Required Packages
import pandas as pd
import numpy as np
```

```python
#Import NFL Schedule Data
NFL2017url='https://raw.githubusercontent.com/holden-herrell/IST718_GROUP_PROJECT/master/DATASETS/NFL%20DATA/NFL2017.csv'
NFL2017=pd.read_csv(NFL2017url, engine='python', error_bad_lines=False)
NFL2017df=pd.DataFrame(NFL2017)
print(NFL2017df)

NFL2018url='https://raw.githubusercontent.com/holden-herrell/IST718_GROUP_PROJECT/master/DATASETS/NFL%20DATA/NFL2018.csv'
NFL2018=pd.read_csv(NFL2018url, engine='python', error_bad_lines=False)
NFL2018df=pd.DataFrame(NFL2018)
print(NFL2018df)

NFL2019url='https://raw.githubusercontent.com/holden-herrell/IST718_GROUP_PROJECT/master/DATASETS/NFL%20DATA/NFL2019.csv'
NFL2019=pd.read_csv(NFL2019url, engine='python', error_bad_lines=False)
NFL2019df=pd.DataFrame(NFL2019)
print(NFL2019df)
```

```python
#Add Years to Each Season
NFL2017df['Year']=2017
print(NFL2017df)
NFL2018df['Year']=2018
print(NFL2018df)
NFL2019df['Year']=2019
print(NFL2019df)
```

```python
#Adjust year for playoff months in next year
NFL2017df.loc[NFL2017df['Date'].str.contains('January'), ['Year']] = 2018
NFL2017df.loc[NFL2017df['Date'].str.contains('February'), ['Year']] = 2018
print(NFL2017df)
NFL2018df.loc[NFL2018df['Date'].str.contains('January'), ['Year']] = 2019
NFL2018df.loc[NFL2018df['Date'].str.contains('February'), ['Year']] = 2019
print(NFL2018df)
NFL2019df.loc[NFL2019df['Date'].str.contains('January'), ['Year']] =2020
NFL2019df.loc[NFL2019df['Date'].str.contains('February'), ['Year']] = 2020
print(NFL2019df)
```

```python
#Combine all NFL Seasons
NFLSchedule=NFL2017df.append(NFL2018df, sort=False, ignore_index=True)
NFLSchedule=NFLSchedule.append(NFL2019df, sort=False, ignore_index=True)
#pd.options.display.max_rows = 999
print(NFLSchedule)
```

```python
#Rename unnamed columns
NFLSchedule.rename(columns={"Unnamed: 5":"At","Unnamed: 7":"BoxScore"},inplace=True)
print(NFLSchedule)
```

```python
#Add HomeTeam
NFLSchedule['HomeTeam']=NFLSchedule['Winner/tie']
```

```python
#Fix HomeTeam to account for @
NFLSchedule.loc[NFLSchedule['At']=='@', ['HomeTeam']] = NFLSchedule['Loser/tie']
print(NFLSchedule)
```

```python
#Subset for cities in this study
NFL= NFLSchedule[NFLSchedule['HomeTeam'].isin(['Atlanta Falcons','Dallas Cowboys','Houston Texans','Miami Dolphins'])]
print(NFL)
```

```python
#Drop Unneeded columns
NFL=NFL.drop(columns=['Week','Winner/tie','At','Loser/tie','BoxScore','PtsW','PtsL','YdsW','TOW','YdsL','TOL'])
print(NFL)
```
